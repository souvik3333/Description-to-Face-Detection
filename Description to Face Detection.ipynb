```python
from PIL import Image
def crop(image_path, coords, saved_location):
    """
    @param image_path: The path to the image to edit
    @param coords: A tuple of x/y coordinates (x1, y1, x2, y2)
    @param saved_location: Path to save the cropped image
    """
    image_obj = Image.open(image_path)
    cropped_image = image_obj.crop(coords)
    cropped_image.save(saved_location)
#     cropped_image.show()
```

```python

from watson_developer_cloud import VisualRecognitionV3

def imgProperty(image_path):
    visual_recognition = VisualRecognitionV3(
    '2018-03-19',
    iam_apikey='QUp3lCzBg79S96jUGaMPzYy83xQeZWEod5ELq2FF6UnQ')
    with open(image_path, 'rb') as images_file:
        classes = visual_recognition.classify(
            images_file,
            threshold='0',
        classifier_ids='default').get_result()
#     print(json.dumps(classes, indent=2))
    return classes
```

```python

def faceLocationAndProperty(image_path, saved_location,counter,data):
    with open(image_path, 'rb') as images_file:
        faces = visual_recognition.detect_faces(images_file).get_result()
        if(len(faces['images'][0]['faces'])==0):
            return counter,data
#     print(json.dumps(faces, indent=2))
#     print(len(faces['images'][0]['faces']))
    for i in range(len(faces['images'][0]['faces'])):
        tface=faces['images'][0]['faces'][i]
        fpath=saved_location+str(counter)+"_"+str(i)+".jpg"
        crop(image_path, (tface['face_location']['left'], tface['face_location']['top'], tface['face_location']['left']+tface['face_location']['width'], tface['face_location']['top']
                         +tface['face_location']['height']), fpath)
        data[fpath]=imgProperty(fpath)
        data[fpath]["images"][0]['gender']=tface['gender']['gender_label']
        data[fpath]["images"][0]['maxAge']=tface['age']['max']
        data[fpath]["images"][0]['minAge']=tface['age']['min']
        data[fpath]["images"][0]['ratio']=(tface['face_location']['height'])/(tface['face_location']['width'])
#         t=pd.Series([fpath])
    return counter+1,data
        
```

```python
raw_documents = []
for i in range(0,len(df)):
    raw_documents.append(df.iloc[i]["property"])
print("Number of documents:",len(raw_documents))
# print(raw_documents)
```

```python
gen_docs = [[w.lower() for w in word_tokenize(text)] 
            for text in raw_documents]
print(gen_docs)
```

```python
dictionary = gensim.corpora.Dictionary(gen_docs)
# print(dictionary[5])
# print(dictionary.token2id['color'])
print("Number of words in dictionary:",len(dictionary))
for i in range(len(dictionary)):
    print(i, dictionary[i])
```

```python
corpus = [dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]
print(corpus)
```

```python
tf_idf = gensim.models.TfidfModel(corpus)
print(tf_idf)
s = 0
for i in corpus:
    s += len(i)
print(s)
```

```python
sims = gensim.similarities.Similarity('match',tf_idf[corpus],
                                      num_features=len(dictionary))
print(sims)
print(type(sims))
```

```python
query_doc = [w.lower() for w in word_tokenize(input_data.iloc[0]["property"])]
print(query_doc)
query_doc_bow = dictionary.doc2bow(query_doc)
print(query_doc_bow)
query_doc_tf_idf = tf_idf[query_doc_bow]
print(query_doc_tf_idf)
```

```python
if len(query_doc_tf_idf)==0:
    print("No match found")
    exit(0)
```

```python
result_array=sims[query_doc_tf_idf]
result_array
```
